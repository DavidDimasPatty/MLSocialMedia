```python
#import package
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#LOAD DATA
```

```python
#DATA BBC
import pandas as pd
myFile = pd.read_csv('PostData/bbc.csv', sep=',', header=None, skiprows=1)

# Mengambil teks dari kolom pertama dari setiap baris dan mengonversinya menjadi list
bbc = myFile.iloc[:, 0].tolist()

# Menampilkan list teks
print(bbc)
```

```python
#DATA CNN
myFile = pd.read_csv('PostData/cnn.csv', sep=',', header=None, skiprows=1)

# Mengambil teks dari kolom pertama dari setiap baris dan mengonversinya menjadi list
cnn = myFile.iloc[:, 0].tolist()

# Menampilkan list teks
print(cnn)
```

```python
#DATA AlJAZER
myFile = pd.read_csv('PostData/al_jazeera.csv', sep=',', header=None, skiprows=1)

# Mengambil teks dari kolom pertama dari setiap baris dan mengonversinya menjadi list
aljazer = myFile.iloc[:, 0].tolist()

# Menampilkan list teks
print(aljazer)
```

```python
#DATA reuters
myFile = pd.read_csv('PostData/reuters.csv', sep=',', header=None, skiprows=1)

# Mengambil teks dari kolom pertama dari setiap baris dan mengonversinya menjadi list
reuters = myFile.iloc[:, 0].tolist()

# Menampilkan list teks
print(reuters)
```

```python
#DATA sentimen
myFile = pd.read_csv('PostData/sentiment_analysis.csv', sep=',', header=None, skiprows=1)

# Mengambil teks dari kolom pertama dari setiap baris dan mengonversinya menjadi list
sentimen = myFile.iloc[:, 4].tolist()

# Menampilkan list teks
print(sentimen)
```

```python
#DATA taylor
myFile = pd.read_csv('PostData/TaylorSwift13.csv', sep=',', header=None, skiprows=1)

# Mengambil teks dari kolom pertama dari setiap baris dan mengonversinya menjadi list
taylor = myFile.iloc[:, 0].tolist()

# Menampilkan list teks
print(taylor)
```

```python
#Data Modeling
```

```python
ALLDATA=[]
ALLDATA+=bbc
ALLDATA+=cnn
ALLDATA+=aljazer
ALLDATA+=reuters
ALLDATA+=sentimen
ALLDATA+=taylor
#print(ALLDATA)
cleaned_data = [text for text in ALLDATA if isinstance(text, str)]
print(cleaned_data)
```

```python
#Process dengan TF - IDF(Seberapa Kata Sering Muncul) dan K-Means (bagi-bagi kelompok)
```

```python
# Inisialisasi TF-IDF Vectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(cleaned_data)

# Inisialisasi model K-Means
kmeans_model = KMeans(n_clusters=2, random_state=42)
kmeans_model.fit(X)

# Prediksi cluster untuk setiap posting
clusters = kmeans_model.predict(X)

# Hitung skor siluet untuk evaluasi model
silhouette_avg = silhouette_score(X, clusters)
print("Silhouette Score:", silhouette_avg)

# Contoh posting baru
new_post = ["I Like Taylor"]

# Transformasi posting baru menjadi vektor TF-IDF
new_post_vector = vectorizer.transform(new_post)

# Prediksi cluster untuk posting baru
new_post_cluster = kmeans_model.predict(new_post_vector)

# Cari posting dalam cluster yang sama dengan posting baru
similar_posts_indices = [i for i, cluster in enumerate(clusters) if cluster == new_post_cluster]

# Print hasil rekomendasi
print("Posting yang direkomendasikan:")
for idx in similar_posts_indices:
    print(cleaned_data[idx])
```
